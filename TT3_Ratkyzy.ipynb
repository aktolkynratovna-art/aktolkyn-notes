```python
# 1 тапсырма
data_rows = []
with open("Salaries.csv", "r", encoding="utf-8") as file:
    for i, line in enumerate(file):
        if i == 0:
            continue
        if i == 11:
            break
        data_rows.append(line.strip().split(","))
print(len(data_rows))
print(data_rows[0])

```

```python
# 2 тапсырма
name_index = 1
title_index = 2
first_three_names = tuple(row[name_index] for row in data_rows[:3])
employee_titles = {row[name_index]: row[title_index] for row in data_rows[:3]}
print("first_three_names =", first_three_names)
print("employee_titles =", employee_titles)
```

```python
# 3 тапсырма
basepay_index = 3
base_pays_str = [row[basepay_index] for row in data_rows]
print("base_pays_str =", base_pays_str)
```
