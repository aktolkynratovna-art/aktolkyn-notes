try-except көмегімен пайдаланушы енгізетін екі санды бөліңіз (қателерді өңдеңіз)

```python
try:
    a = float(input("Бірінші санды енгізіңіз: "))
    b = float(input("Екінші санды енгізіңіз: "))
    result = a / b
    print("Нәтиже:", result)
except ValueError:
    print("Қате: Сан емес мәтін енгізілді.")
except ZeroDivisionError:
    print("Қате: Нөлге бөлуге болмайды.")
except Exception as e:
    print("Белгісіз қате:", e)
```

Rectangle деген класс құрып, оның ауданы мен периметрін есептейтін әдістерін жазыңыз.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

w = float(input("Енін енгізіңіз: "))
h = float(input("Биіктігін енгізіңіз: "))

rect = Rectangle(w, h)

print("Ауданы:", rect.area())
print("Периметрі:", rect.perimeter())
```

NumPy арқылы 10 элементтен тұратын кездейсоқ массив жасап, оның орта мәнін, максимум және минимумын табыңыз.

```python
import numpy as np

arr = np.random.randint(0, 101, size=10)
print("Массив:", arr)

a = np.mean(arr)
print("Орта мән:", a)


max_b = np.max(arr)
print("Максимум:", max_b)

min_c = np.min(arr)
print("Минимум:", min_c)
```

```python
import random

arr = []
for i in range(10):
    arr.append(random.randint(0, 100))

print("Массив:", arr)


sum_a = 0
max_b = arr[0]
min_c = arr[0]

for i in range(10):
    num = arr[i]
    sum_a += num
    if num > max_b:
        max_b = num
    if num < min_c:
        min_c = num

s = sum_a/ len(arr)

print("Орта мән:", s)
print("Максимум:", max_b)
print("Минимум:", min_c)

```
